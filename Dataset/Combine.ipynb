Combine individual time series into multivariate time series

```python
import pandas as pd
import os
from pathlib import Path
```

```python
TIME_SERIES_DIR = "./Time Series"
time_series = []
```

```python
files = os.listdir(TIME_SERIES_DIR)
csv_files = [file for file in files if file.endswith('.csv')]

# For every csv file in Time Series folder, load into dataframe and add to list
for csv_file in csv_files:
    file_path = os.path.join(TIME_SERIES_DIR, csv_file)
    df = pd.read_csv(file_path, header=[0,1], index_col = 0)
    df.index.name = Path(file_path).stem
    df.index = pd.to_datetime(df.index)
    df.index.freq = "D"
    time_series.append(df)
    print(f"{Path(file_path).stem} contains {df.shape[0]} entries.")
```

```python
# Use inner join to merge time series so date range contains data with all indexes
# (only select entries where all timeseries has data)

df = time_series[0]
for ts in time_series[1:]:
    df = pd.merge(df, ts, how="inner", left_index=True, right_index=True)

print(f"Merged dataframe contains {df.shape[0]} entries.")
```

```python
df = df.sort_index(axis=1)
df.head()
```

```python
df.to_csv("time_series.csv", index_label=False)
```
